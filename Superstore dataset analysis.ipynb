<h2> 5.1 Analysis of Numerical Quantities </h2>

<p> Lets analyse the distribution of the Quantity, Sales, Profit and Discount in the dataset </p>

```python
# filter the dataset to store the numerical data in a separate dataframe
df_num = df[["Sales", "Quantity", "Discount", "Profit"]]
```

```python
# Maximum values for each of the numerical quantities
df_num.max()
```

```python
# Minimum values for each of the numerical quantities
df_num.min()
```

```python
# Plot the Histogram, Violin-plot and Box-plot graphs for each of the numerical quantities

# iterate through the numerical quantity dataframe and plot each graph
for column in df_num:
    plt.figure(figsize =(14,5)) # size of figure
    
    plt.subplot(1,3,1)
    sns.histplot(data = df_num , x = column, color = 'darkgreen', bins = 14) # histogram using seaborn
    plt.title(f'Histogram of {column}', size=10)
    
    plt.subplot(1,3,2)
    sns.violinplot(x = df[column]) # violin-plot using seaborn
    plt.title(f'Violin-plot of {column}', size=10)
    
    plt.subplot(1,3,3)
    sns.boxplot(data = df_num , x = column, color = 'purple') # box-plot using seaborn
    plt.title(f'Box-plot of {column}', size=10)
    plt.tight_layout() # tight layout
```

<h3> Observations: </h3>
<p> <b> Sales: </b> The Sales distribution is skewed right, with outliers on the right side of the plots. The product sales of zero has the highest frequency in the dataset. </p>

<p> <b> Quantity: </b> From this we see that the Quantity distribution is skewed right, with outliers on the right side of the plots. The product quantities between two and four have the highest frequency in the dataset. </p>

<p> <b> Discount: </b> The Discount distribution seems to be multimodal since shows two peaks in the histogram and violin-plot, with outliers on the right side of the plots. The product discount of zero has the highest frequency in the dataset. </p>

<p> <b> Profit: </b> <br> The Profit has a normal dirstribution but is difficult to analyse in the plots as the values are very small. </p>

<h3> 5.2.1 Heat Map </h3>

<p> <b> Observation: </b> We see that compared to other numerical quantities, the sales and profit are correlated to a small extent as the closer the values are to one, the more positively correlated they are. The negative value implies negative correlation, zero implies no relation and higher the value means more the correlation of the quantities. </p>

<p> <b> Observation: </b> The sales and profit are somewhat correlated to each other based on the pair plot shown above as it shows a linear graph between the two quantities. </p>

<h2>9. What are the Sub-Category sales in each of the Customer Segments? </h2>

<h2> 16. What is the trend of the Sales and Profit over the years? </h2>

```python
# Group the average Sales and Profit by the order date
df_avg_time = df.groupby(["Sales","Profit"])["Order Date"].mean()

plt.figure(figsize=(15,5)) # figure size

sns.lineplot(x=df["Order Date"], y= df["Sales"],data=df_avg_time) # plot average Sales over time

sns.lineplot(x=df["Order Date"], y= df["Profit"],data=df_avg_time) # plot average Profit over time

plt.title("Average Sales and Profit over time (2014-2018)") # plot title
```

<h3> 16.1 Sales </h3>

```python
# indexing by order date
df_time = df.set_index('Order Date')

# using resample from pandas to set the resampled order date as the index for the plot
df_time["Sales"].resample("M").sum().plot()

plt.title("Total Sales over time") # plot title
```

<h3> 16.1.1 Lets Analyse this trend further </h3>

```python
# using seasonal_decompose from statsmodels.tsa.seasonal module to plot the decomposition chart of the Sales over time
sales = seasonal_decompose(df_time["Sales"].resample("M").sum(), model='additive')

# draw all plots of the sales
sales.plot()
plt.show()
```

<p> <b> Observation: </b> From the Trend plot in the decomposition chart, we see that the sales gradually increases from 2014 to 2017. The seasonal plot shows that the sales keeps fluctuating every year with peaks in the months of January, November and December of each year. </p>

<h3> 16.2 Profit </h3>

```python
# using resample from pandas to set the resampled order date as the index for the plot
df_time["Profit"].resample("M").sum().plot()

plt.title("Total Profit over time") # plot title
```

<p> <b> Observation: </b> We see that the Profit peaked in January of 2017, and slowly decreases toward the end of the year.

<h3>16.2.2 Further Analysis of Profit over time</h3>

```python
# using seasonal_decompose from statsmodels.tsa.seasonal module to plot the decomposition chart of the Profit over time
profit = seasonal_decompose(df_time["Profit"].resample("M").sum(), model='additive')

# draw all plots of the sales
profit.plot()
plt.show()
```

<p> <b> Observation: </b> From the Trend plot in the decomposition chart, we see that the Profit has an increasing trend from 2014 to 2017. The seasonal plot shows that the Profit fluctuates with peaks in the months of January, November and December, similar to that of Sales. </p>
